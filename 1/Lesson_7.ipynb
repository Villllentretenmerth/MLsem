## Парсинг

```python
import re

html_text = '''<html>
                  <body>
                    <h1>Мой заголовок</h1>
                    <p>Мой абзац.</p>
                  </body>
                </html>'''
# Поиск всех заголовков h1
pattern = r'<h1>(.*?)</h1>'
titles = re.findall(pattern, html_text)
titles # -> ['Мой заголовок']
```

```python
import re

html = '<a href="https://example.com">Link</a>'
# Упрощенное регулярное выражение для демонстрации
pattern = r'<a\s+href="(.*?)">'
links = re.findall(pattern, html)
print(links)  # Вывод: ['https://example.com']
```

### JSON

```python
import json

data = {
    'name': 'Alice',
    'age': 30,
    'cities': ['London', 'Paris']
}

json_string = json.dumps(data, indent=2)  # indent для красивого форматирования
print(json_string) # {"name": "Alice", ...}
```

## API

```python
url = 'http://api.openweathermap.org/data/2.5/weather?q=London,uk&APPID=e86ee0857405675859d1fa65034b600e'

response = requests.get(url)
data = response.json()

print(f"Температура: {data['main']['temp']}K")
print(f"Погода: {data['weather'][0]['description']}")
```

```python
import requests

#url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}"
url = f"http://api.openweathermap.org/data/2.5/weather"

api_key = "e86ee0857405675859d1fa65034b600e"
city = "London"
params = {'q': city,
          'appid': api_key,
          'units': 'metric'
         }

response = requests.get(url, params=params)
data = response.json()

print(f"Температура: {data['main']['temp']} C")
print(f"Погода: {data['weather'][0]['description']}")
```
